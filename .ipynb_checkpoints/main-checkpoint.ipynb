```python
coinciden = True
for i in lista_nombres:
    if df[df.nombre_personal== i].ssexo.nunique() != 1 or df[df.nombre_personal== i].rol_personal.nunique() != 1:
        coinciden = False
    if not coinciden:
        print('No coincide en ', i)
        break
print('Coinciden: ', coinciden)
```
